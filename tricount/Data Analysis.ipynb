```python
from forex_python.converter import CurrencyRates, C

def get_all_currencies():
    c = CurrencyRates()
    currencies = c.get_rates('USD')  # Get exchange rates relative to USD
    currency_list = [(code, code + f" ({symbol})") for code, symbol in currencies.items()]
    return currency_list

# Call the function to get the list of all currencies
all_currencies = get_all_currencies()

for currency in all_currencies:
    print(currency)
```

```python

def get_all_currency_symbols():
    c = CurrencyRates()
    currencies = c.get_rates('USD')  # Get exchange rates relative to USD
    currency_symbols = {code: symbol for code, symbol in currencies.items()}
    return currency_symbols

# Call the function to get the dictionary of currency symbols
all_currency_symbols = get_all_currency_symbols()

for code, symbol in all_currency_symbols.items():
    print(f"{code}: {symbol}")
```

```python
import requests

def get_currency_info():
    base_url = "https://openexchangerates.org/api/currencies.json"
    app_id = "YOUR_APP_ID"  # Replace with your Open Exchange Rates app ID

    response = requests.get(f"{base_url}?app_id={app_id}")
    currency_info = response.json()

    return currency_info

# Call the function to get currency information
all_currency_info = get_currency_info()

for code, name in all_currency_info.items():
    print(f"{code}: {name}")
```
